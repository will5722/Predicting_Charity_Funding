## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME', 'USE_CASE'], axis=1)
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_count = application_df.APPLICATION_TYPE.value_counts()
app_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_count[app_count< 600].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_count = application_df.CLASSIFICATION.value_counts()
class_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_count_reduced = application_df.CLASSIFICATION.value_counts().loc[lambda x : x>1]
class_count_reduced
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_count[class_count< 1900].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
cols_to_convert = application_df.dtypes[application_df.dtypes == "object"].index
application_df_dummies = pd.get_dummies(application_df, columns=cols_to_convert, drop_first=True)
application_df_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_dummies.drop(['IS_SUCCESSFUL'], axis=1)
y = application_df_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

# First hidden layer
# attempt #1: nn.add(tf.keras.layers.Dense(units=20, activation='relu', input_dim=len(X_train_scaled[0])))
# attempt #2: nn.add(tf.keras.layers.Dense(units=60, activation='relu', input_dim=len(X_train_scaled[0])))
nn.add(tf.keras.layers.Dense(units=80, activation='relu', input_dim=len(X_train_scaled[0])))


# Second hidden layer
# attempt #1: nn.add(tf.keras.layers.Dense(units=20, activation='relu'))
# attempt #2: nn.add(tf.keras.layers.Dense(units=60, activation='relu'))
nn.add(tf.keras.layers.Dense(units=70, activation='relu'))

# Adding third hidden layer on third attempt
nn.add(tf.keras.layers.Dense(units=60, activation='sigmoid'))


# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))



# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Callback to save weights
#from keras.callbacks import Callback
callback = tf.keras.callbacks.ModelCheckpoint(filepath="Checkpoints/Checkpoints.{epoch}.hdf5", verbose=1, save_weights_only=True, save_freq="epoch", period=5)

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[callback])
```

```python
# Evaluate the model using the test data
# Attempt #1: Loss: 0.5681890845298767, Accuracy: 0.7197667360305786 
# Attempt #2: Loss: 0.5754483938217163, Accuracy: 0.7194169163703918, changed number of units in 1st & 2nd layers
# Attempt #3: Loss: 0.5903772115707397, Accuracy: 0.7188338041305542, changed number of units in 1st & 2nd layers, 
# added 3rd layer, deleted USE_CASE column, slightly increased values for app_count & class_count.
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
